# Обработка изображений

```python
import numpy as np
import matplotlib.pylab as plt

%matplotlib inline
```

```python
im = plt.imread("Lenna.png")
im.shape
```

```python
im[0][0]
```

```python
def plti(im, h=8, **kwargs):
    """
    Helper function to plot an image.
    """
    y = im.shape[0]
    x = im.shape[1]
    w = (y/x) * h
    plt.figure(figsize=(w,h))
    plt.imshow(im, interpolation="none", **kwargs)
    plt.axis('off')

plti(im)
```

```python
plti(im[20:300, 0:300, :])
```

```python
plti(im[:, :, 1])
```

```python
grey = np.zeros((im.shape[0], im.shape[1])) # init 2D numpy array

for rownum in range(len(im)):
    for colnum in range(len(im[rownum])):
        grey[rownum][colnum] = np.average(im[rownum][colnum])
```

```python
grey[0][0]
```

```python
im[0][0]
```

```python
plti(grey, cmap='Greys')
```
