Here are two things we need to do:

1) Many of the Instagram users in this dataset have apparently were using automatic tag generation apps that add long sets of generic tags in English. We want to delete such posts from the dataset.

2) Many of the users who have lots of posts are from businesss. They can be identified in a simple way - all locatons of their posts are the same. We also want to delete these posts.

I suggest that to indicate results of the filtering, you add a column to the data that indicates if a particular image is a) has auto-generated English tags, and/or b) likely to be from a company.

1. Wrote more than 1 post.

```python
spam_users = set()
for user, d in tqdm_notebook(data.groupby("username")):
    if len(d) > 1:
        hours = (max(d["local_time"]) - min(d["local_time"])).total_seconds() / 3600
        if (len(d["lat"].unique()) == 1) and (len(d["lon"].unique()) == 1) and (hours >= 1):
            spam_users.add(user)
len(spam_users)
```

```python
spam_users_2 = set()
for user, d in tqdm_notebook(data.groupby("username")):
    if len(d) > 1:
        if (len(d["lat"].unique()) == 1) and (len(d["lon"].unique()) == 1):
            spam_users_2.add(user)
len(spam_users_2)
```

```python
data[data["username"].isin(spam_users)].sort_values("username")
```

Business and spam accounts can alse be detected by number of posts/number of unique tags thay used. It's  suspicious if user writes a lot of posts with the same tags.

```python
users = {}
for name, group in tqdm_notebook(data[~data["username"].isin(spam_users)].groupby("username")):
    users[name] = {}
    tags_set = set()
    for t in group["tags"]:
        tag_set = set(str(t).split(","))
        tags_set.update(tag_set)
    users[name]["used_tags_num"] = len(tags_set)
    users[name]["posts_num"] = len(group["tags"])
```

```python
users_df = pd.DataFrame.from_dict(users, orient="index")
```

```python
ax = users_df.plot(kind='scatter', x="posts_num", y="used_tags_num")
ax.set_xlim([0, 500])
ax.set_ylim([0, 500])
```

```python
buisiness_acc = set(["apriori_salon", "bar_nebar", "ginzaprojectspb", "hashtag.tsvety", "keratinka_spb", "nails_dj_anna_li", "nata_keratin", "street_dance_spb"])
```

```python
spam_users = spam_users.union(buisiness_acc)
```

```python
data["buisiness_account"] = data["username"].isin(spam_users)
```

```python
data["buisiness_account_2"] = data["username"].isin(spam_users_2)
```

```python
tag_list = []
for l in data[data["tags"].notnull() & ~data["buisiness_account"]]["tags"].str.split(","):
    tag_list += l
```

```python
len(tag_list)
```

```python
c = Counter(tag_list)
```

```python
c.most_common()
```

```python
vsco_tags = []
for tag, count in c.most_common():
    if tag.startswith("vsco") or tag.endswith("vsco"):
        print(tag)
        vsco_tags.append(tag)
```

```python
list_of_trashtags = set(vsco_tags + ["instasize", "маникюрспб", "ногтиспб", "шеллакспб", "дизайнногтей", "гельлакспб", "instacollage", "shellacspb", "webstagram", "instago", "педикюрспб", "наращиваниересниц", "кератиновоевыпрямлениеспб", "кератинспб", "latergram", "lianail", "наращиваниересницспб", "instapic", "шилакспб"])
len(list_of_trashtags)
```

```python
def remove_trash_tags(string_tags):
    if isinstance(string_tags, str):
        tags_list = set(string_tags.split(","))
        good_tags = set(tags_list).difference(list_of_trashtags)
        good_tags_string = ",".join(list(good_tags))
        if good_tags_string:
            return good_tags_string
        else:
            return np.nan
    else:
        return np.nan
```

```python
def trash_tag_ratio(string_tags):
    if isinstance(string_tags, str):
        tags_list = set(string_tags.split(","))
        intersecter = tags_list.intersection(list_of_trashtags)
        ratio = len(intersecter)/len(tags_list)
        return ratio
    else:
        return np.nan
```

```python
data["trashtags_ratio"] = data["tags"].apply(trash_tag_ratio)
```

```python
data[data["trashtags_ratio"] > 0.5]
```

```python
data["buisiness_account"] = data["buisiness_account"].astype(int)
data["buisiness_account_2"] = data["buisiness_account_2"].astype(int)
```
